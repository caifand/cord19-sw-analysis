```python
import pandas as pd
import numpy as np
import json
```

```python
pd.set_option('display.max_columns', 800)
pd.set_option('display.max_rows', 200)
```

```python
with open('documents.json', encoding='utf8') as f:
    d = json.load(f)
```

```python
docs_full = pd.json_normalize(d)
docs_full.head(3)
```

```python
docs = docs_full[['_key', 'metadata.DOI', 'metadata.cord_sha', 'metadata.title', 'metadata.subtitle', \
                              'metadata.container-title', 'metadata.short-container-title', \
                              'metadata.type', 'metadata.subject', 'metadata.source', 'metadata.journal-issue.issue', \
                              'metadata.issued.date-parts', 'metadata.publisher', 'metadata.link']]
```

```python
docs = docs.rename(columns={'_key': 'doc_key', 'metadata.DOI': 'DOI', 'metadata.cord_sha': 'cord_sha', \
                             'metadata.title': 'title', 'metadata.subtitle': 'subtitle', \
                             'metadata.container-title': 'journal_title', \
                             'metadata.short-container-title': 'journal_short_title', \
                             'metadata.type': 'document_type', 'metadata.subject': 'journal_subject',
                             'metadata.source': 'metadata_source', \
                             'metadata.journal-issue.issue': 'journal_issue', 'metadata.publisher': 'publisher',\
                             'metadata.issued.date-parts': 'issue_dates', \
                             'metadata.abstract': 'abstract', 'metadata.link': 'link'})
```

```python
docs['journal_title'] = docs['journal_title'].apply(lambda x: ','.join(map(str, x)))
docs['journal_short_title'] = docs['journal_short_title'].apply(lambda x: ','.join(map(str, x)))
docs['title'] = docs['title'].apply(lambda x: ','.join(map(str, x)))
docs['subtitle'] = docs['subtitle'].apply(lambda x: ','.join(map(str, x)))
```

```python
docs.to_csv('data/docs.csv', index=False)
```

```python
with open('annotations.json', encoding='utf8') as f:
    a = json.load(f)
```

```python
annos = pd.json_normalize(a)
annos.head(3)
```

```python
annos = annos[['_key', 'context', 'document.$oid', 'software-name.offsetStart', 'software-name.normalizedForm', \
              'software-name.offsetEnd', 'software-name.rawForm', 'references', 'version.offsetStart', \
              'version.normalizedForm', 'version.offsetEnd', 'version.rawForm', 'publisher.offsetStart', \
              'publisher.normalizedForm', 'publisher.offsetEnd', 'publisher.rawForm', \
              'url.offsetStart', 'url.normalizedForm', 'url.offsetEnd', 'url.rawForm']]
```

```python
annos = annos.rename(columns={'_key': 'anno_key', 'document.$oid':'doc_key'})
annos.head(3)
```

```python
annos_docs_jif = annos.merge(docs_jif, how='left', on='doc_key')
```

```python
annos_docs_jif['issue_year'] = annos_docs_jif['issue_year'].astype("int")
```

```python
annos_docs_jif['issue_year'].unique()
```

```python
annos_docs_jif = annos_docs_jif[annos_docs_jif['issue_year'] > 2015]
```

```python
annos_docs_jif['doc_key'].nunique()
```

```python
annos_docs_jif = annos_docs_jif[annos_docs_jif['document_type'] == 'journal-article']
```

```python
annos_docs_jif['doc_key'].nunique()
```

```python
annos_docs_jif[annos_docs_jif['rank'] <= 10]
```

```python
annos_docs_jif.loc[annos_docs_jif['rank'] <= 10, 'stratum'] = "[1,10]"
annos_docs_jif.loc[(annos_docs_jif['rank'] > 10) & (annos_docs_jif['rank'] <= 100), \
                        'stratum'] = "[11,100]"
annos_docs_jif.loc[(annos_docs_jif['rank'] > 100) & (annos_docs_jif['rank'] <= 1000), \
                        'stratum'] = "[101,1000]"
annos_docs_jif.loc[(annos_docs_jif['rank'] > 1000) & (annos_docs_jif['jif'].notna()), 'stratum'] = "[1001,12895]"
annos_docs_jif.loc[annos_docs_jif['jif'].isna(), 'stratum'] = "[12896,)"
```

```python
annos_docs_jif.groupby(by='stratum').size()
```

```python
annos_docs_jif.shape
```

```python
stratum_type = pd.CategoricalDtype(categories=['[1,10]', '[11,100]', '[101,1000]', '[1001,12895]', '[12896,)'], ordered=True)
```

```python
annos_docs_jif['stratum'] = annos_docs_jif['stratum'].astype(stratum_type)
annos_docs_jif['stratum'] = annos_docs_jif['stratum'].cat.rename_categories({"[12896,)": "No Impact Factor"})
```

```python
doc_mention_density = pd.DataFrame(annos_docs_jif.groupby(by='doc_key').size()).reset_index().set_axis(['doc_key', 'mention_density'], axis=1)
doc_mention_density.sort_values(by='mention_density', ascending=False).head(3)
```

```python
annos_docs_jif = annos_docs_jif.merge(doc_mention_density, how='left', on='doc_key')
annos_docs_jif.head(3)
```
